## Import Modules

```python
import matplotlib.pyplot as plt
import PIL
import pytesseract
import re
%matplotlib inline
```

```python
# prerequisites
# brew install tesseract (install in system terminal)
# !pip install pytesseract
# install desktop version of pytesseract
```

## Load the image

```python
img = PIL.Image.open('test.JPG')
plt.imshow(img)
```

Set the path to the Tesseract executable (you may not need this if it's in your system PATH)

```python
pytesseract.pytesseract.tesseract_cmd = '/usr/local/bin/tesseract'
```

## Convert Image to Text

```python
text_data = pytesseract.image_to_string(img.convert('RGB'), lang='eng')
```

```python
print(text_data)
```

## Extract Specific Fields

```python
m = re.search("Name: (\w+)", text_data)
name = m[1]
name
```

```python
m = re.search("Start Date: (\S+)", text_data)
start_date = m[1]
start_date
```

```python
m = re.search("Geo-Coordinates: (\S+)", text_data)
coordinates = m[1]
coordinates
```
